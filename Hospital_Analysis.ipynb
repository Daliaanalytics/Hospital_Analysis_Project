```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= {
    'Patient_ID': range(1,21),
    'Age':[25,34,45,67,23,56,78,43,32,61,29,54,71,38,49,63,27,52,44,35],
    'Gender':['M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F','M','F'],
    'Diesease': ['Flu','Diabetes','Heart','Flu','Fracture','Diabetes','Heart','Flu','Fracture','Heart',
                'Flu','Diabetes','Heart','Fracture','Diabetes','Heart','Flu','Fracture','Diabetes','Flu'],
    'Department': ['General','Endocrine','Cardiology','General','Orthopedic','Endocrine','Cardiology',
                   'General','Orthopedic','Cardiology','General','Endocrine','Cardiology','Orthopedic',
                   'Endocrine','Cardiology','General','Orthopedic','Endocrine','General'],
    'Days_Stayed': [2,5,7,3,4,6,8,2,5,9,2,5,8,4,6,9,2,4,5,3],
    'Treatment_Cost':[500,2000,5000,800,3000,2500,6000,600,3500,7000,500,2000,6500,3000,2500,7000,500,3000,2000,700],
    'Recovered':['Yes','Yes','No','Yes','Yes','No','Yes','Yes','No','Yes','Yes','Yes','No','Yes','Yes','No','Yes','Yes','Yes','Yes']
}

df= pd.DataFrame(data)
```

```python
df['Diesease'].value_counts()
```

```python
diesease_counts= df['Diesease'].value_counts()
plt.pie(diesease_counts, labels= diesease_counts.index,autopct='%1.1f%%')
plt.title('Counts of Dieseases')
plt.show()
```

```python
bins= [0,30,50,100]
labels=['Young(0-30)', 'Middle(31-50)', 'Senior(+51)']
df['Age_Group']= pd.cut(df['Age'], bins=bins, labels=labels)
df['Age_Group'].value_counts(sort=False)
```

```python
x=df['Age_Group'].value_counts(sort=False)
plt.pie(x, labels= x.index, autopct='%1.1f%%')
plt.title('Age Group Chart')
plt.show()
```

```python
sns.boxplot(data=df, x='Diesease', y='Age')
plt.title('Age Distribuation by Diesease')
plt.xlabel('Diesease')
plt.ylabel('Age')
plt.show()
```

```python
df.groupby('Diesease')['Days_Stayed'].mean().sort_values(ascending=False)
```

```python
x=df.groupby('Diesease')['Days_Stayed'].mean().sort_values(ascending=False)
plt.bar(x.index, x.values)
plt.title('Average Length of Stay per Diesease')
plt.xlabel('Diesease')
plt.ylabel('Days_Stayed')
plt.show()
```
